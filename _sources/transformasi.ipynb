# TRANSFORMASI MATRIKS

- Transformasi Matriks

Transformasi matriks adalah metode menggunakan matriks untuk mengubah, memodifikasi, atau mengonversi titik-titik dalam ruang vektor. Dalam konteks geometris, transformasi matriks sering diterapkan untuk mengubah bentuk atau posisi objek dalam ruang 2D atau 3D melalui operasi seperti translasi, rotasi, skalasi, refleksi, dan shearing.

1. Skalasi (Scaling):
Mengubah ukuran objek. Matriks skalasi berisi faktor skala untuk sumbu x dan y (atau z dalam 3D).

2. Translasi (Translation):
Menggeser objek dari satu posisi ke posisi lain. Translasi biasanya dilakukan dengan menambahkan vektor translasi ke koordinat objek. Dalam 3D, ini sering dilakukan dengan menggunakan matriks homogen.

- Berikut adalah contoh implementasi tranformasi matriks dalam bahasa program Python

```python
#rotasi transformasi
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menggambar persegi
def plot_square(points, label, color, marker):
    points = np.vstack((points, points[0]))  # Menghubungkan titik terakhir ke titik pertama
    plt.plot(points[:, 0], points[:, 1], marker=marker, color=color, label=label)

# Matriks Rotasi untuk sudut Œ∏ (dalam radian)
def rotation_matrix(theta):
    return np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])

# Titik-titik persegi asli
square_points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1]
])

# Definisikan sudut rotasi (dalam derajat dan konversi ke radian)
theta = np.radians(45)

# Hitung matriks rotasi
rot_matrix = rotation_matrix(theta)

# Terapkan rotasi ke titik-titik persegi
rotated_square = np.dot(square_points, rot_matrix)

# Visualisasi Transformasi Rotasi
plt.figure(figsize=(8, 8))
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)
plt.grid(True, which='both')

plot_square(square_points, "Original Square", 'blue', 'o')
plot_square(rotated_square, "Rotated Square (45 degrees)", 'red', 's')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Rotation Transformation')
plt.legend()
plt.axis('equal')
plt.show()
```

```python
#refleksi transformasi
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menggambar persegi
def plot_square(points, label, color, marker):
    points = np.vstack((points, points[0]))  # Menghubungkan titik terakhir ke titik pertama
    plt.plot(points[:, 0], points[:, 1], marker=marker, color=color, label=label)

# Matriks Refleksi terhadap sumbu x
def reflection_matrix(axis):
    if axis == 'x':
        return np.array([
            [1, 0],
            [0, -1]
        ])
    elif axis == 'y':
        return np.array([
            [-1, 0],
            [0, 1]
        ])
    else:
        raise ValueError("Axis must be 'x' or 'y'")

# Titik-titik persegi asli
square_points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1]
])

# Pilih sumbu refleksi ('x' atau 'y')
axis = 'x'

# Hitung matriks refleksi
ref_matrix = reflection_matrix(axis)

# Terapkan refleksi ke titik-titik persegi
reflected_square = np.dot(square_points, ref_matrix)

# Visualisasi Transformasi Refleksi
plt.figure(figsize=(8, 8))
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)
plt.grid(True, which='both')

plot_square(square_points, "Original Square", 'blue', 'o')
plot_square(reflected_square, f"Reflected Square (axis={axis})", 'red', 's')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Reflection Transformation')
plt.legend()
plt.axis('equal')
plt.show()
```

```python
#shearing transformasi
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk menggambar persegi
def plot_square(points, label, color, marker):
    points = np.vstack((points, points[0]))  # Menghubungkan titik terakhir ke titik pertama
    plt.plot(points[:, 0], points[:, 1], marker=marker, color=color, label=label)

# Matriks Shearing untuk sumbu x dan y
def shearing_matrix(shear_x, shear_y):
    return np.array([
        [1, shear_x],
        [shear_y, 1]
    ])

# Titik-titik persegi asli
square_points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1]
])

# Nilai shearing untuk sumbu x dan y
shear_x = 0.5
shear_y = 0.2

# Hitung matriks shearing
shear_matrix = shearing_matrix(shear_x, shear_y)

# Terapkan shearing ke titik-titik persegi
sheared_square = np.dot(square_points, shear_matrix)

# Visualisasi Transformasi Shearing
plt.figure(figsize=(8, 8))
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)
plt.grid(True, which='both')

plot_square(square_points, "Original Square", 'blue', 'o')
plot_square(sheared_square, "Sheared Square", 'red', 's')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Shearing Transformation')
plt.legend()
plt.axis('equal')
plt.show()
```
