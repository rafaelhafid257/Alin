Matriks adalah sekumpulan bilangan yang disusun berdasarkan baris dan kolom, serta ditempatkan di dalam tanda kurung.

## operasi matriks
### penjumlahan
 penjuamlahan matriks dapat dilakukan jika panjang baris dan kolomm kedua matriks sama

cara menjumlahkannya cukup menambahkan matriks pertama dan kedua pada tiap baris dan kolom yang sama

penjumlahan matrik memiliki sifat sifat:

Sifat komutatif

Pada sifat ini berlaku A + B = B + A.

Sifat asosiatif
Pada sifat ini berlaku (A + B) + C = A +(B + C)

Matriks nol

Pada sifat ini berlaku A + 0 = A.

Jika dijumlahkan dengan lawannya akan menghasilkan matriks
nol, yaitu A + (-A) = 0.
### pengurangan
 sama seperti penjumlahan hanya saja idkurangi
### perkalian
perkalian matriks hanya dapat dilakukan jika kolom matriks A sama dengan baris matrik B

matriks dari hasil perkalian matriks akan memiliki jumlah baris
yang sama dengan matrik pertama dan jumlah kolom yang sama dengan matrik kedua

perkalian matrik A x B berbeda dengan perkalian matrik B x A

perkalian matrik memiliki sifat sifat:

Sifat asosiatif

Pada sifat ini, berlaku (A × B) × C= A × (B × C)


Sifat distributif

Pada sifat ini, berlaku A × (B + C) = (A × B) + (A × C)
###transpose matrik
transpose matrik adalah matrik yang posisi baris dan kolomnya bertukar posisi  

transpose matriks biasanya ditandai dengan tanda pangkat t pada matriks tersebut

transpose matriks memiliki sifat sifat:

1. (PT)T = P

2. (P + Q)T = PT + QT

3. (P – Q)T = PT – QT

4. (KP)T = k.PT, di mana k = suatu konstanta

5. (PQ)T = PTQT
