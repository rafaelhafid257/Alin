##Page rank
###algoritma

PageRank adalah algoritma yang dikembangkan oleh Larry Page dan Sergey Brin, pendiri Google, untuk menentukan peringkat halaman web dalam hasil mesin pencari. Algoritma ini digunakan untuk mengukur pentingnya halaman web berdasarkan jumlah dan kualitas tautan yang mengarah ke halaman tersebut. PageRank didasarkan pada gagasan bahwa halaman web yang lebih penting cenderung menerima lebih banyak tautan dari halaman web lainnya.

Cara kerjanya secara singkat adalah sebagai berikut:

```python
av1 = a @ v
print(av1, '\n')

av2 = a @ av1
print(av2, '\n')

av3 = a @ av2
print(av3, '\n')

av4 = a @ av3
print(av4, '\n')

av5 = a @ av4
print(av5, '\n')

av6 = a @ av5
print(av6, '\n')

av7 = a @ av6
print(av7, '\n')

av8 = a @ av7
print(av8, '\n')
```
