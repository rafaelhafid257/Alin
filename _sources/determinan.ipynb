```python
# importing Numpy package
import numpy as np

# creating a 2X2 Numpy matrix
Z = np.array([[1, 0], [0, 1]])
Y = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
C = np.array([[1, 2, 1], [3, 1, 2], [2, 1, 2]])
D = np.array([[3, 1, 2], [1, 2, 1], [2, 1, 2]])
E = np.array([[1, 2, 1], [2, 4, 2], [2, 1, 2]])
F = np.array([[1, 2, 2], [2, 4, 3], [1, 2, 3]])
G = np.array([[6, 2, 4], [2, 4, 2], [4, 2, 4]])
A = np.array([[3, 4], [2, 1]])
I = np.array([[6, 4], [4, 1]])
B = np.array([[6, 8], [2, 1]])
J = np.array([[4, 1], [2, 1]])
K = np.array([[7, 5], [2, 1]])
L = np.array([[4, 2, 1], [0, 1, 4], [0, 0, 2]])


# calculating the determinant of matrix

print(Z)
print("Z",np.round(np.linalg.det(Z)))
print(Y)
print("Y",np.round(np.linalg.det(Y)))
print(C)
print("C",np.round(np.linalg.det(C)))
print(D)
print("D",np.round(np.linalg.det(D)))
print(E)
print("E",np.round(np.linalg.det(E)))
print(F)
print("F",np.round(np.linalg.det(F)))
print(G)
print("G",np.round(np.linalg.det(G)))
print(A)
print("A",np.round(np.linalg.det(A)))
print(I)
print("I",np.round(np.linalg.det(I)))
print(B)
print("B",np.round(np.linalg.det(B)))
print(J)
print("J",np.round(np.linalg.det(J)))
print(K)
print("K",np.round(np.linalg.det(K)))
print(L)
print("L",np.round(np.linalg.det(L)))

```
